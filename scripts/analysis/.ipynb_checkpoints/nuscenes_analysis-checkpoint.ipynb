```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load accuracy .csv and times.csv
time_path = '../../results/mini_val/times.csv'
acc_path = '../../results/accuracy.csv'

def get_max_n(row):
    return max(row['n_dets'],row['n_trks'])

time_df = pd.read_csv(time_path)
time_df['n_max'] = time_df.apply(get_max_n, axis=1)
acc_df = pd.read_csv(acc_path)

# Get max time index
max_ind = time_df['time_ns'].idxmax()
max_mask = time_df.index==max_ind
```

# Plot 1: Linearity in O(n)

```python
# Generate timing plot
n_bins=20
fig, axs = plt.subplots(1, 2, tight_layout=False)
axs[0].scatter(time_df[~max_mask]['n_max'],time_df[~max_mask]['time_ns'],marker='.')
axs[0].set_xlim([0,250])
axs[0].set_title('Raw Tracker Times')
axs[0].set_xlabel('max($n_{tracks},n_{dets}$)')
axs[0].set_ylabel('Tracker Loop Interval $\Delta t$ (ns)')
axs[1].hist(time_df[~max_mask]['time_ns']/time_df[~max_mask]['n_max'],density=True,stacked=True,bins=25) #, bins=n_bins
axs[1].set_title('Normalized Tracker Times')
axs[1].set_ylabel('prob($\\bar{\\tau}$)')
axs[1].set_xlabel('$\\bar{\\tau}$ (nSPOT)')

print(np.mean(time_df[~max_mask]['time_ns']/time_df[~max_mask]['n_max']))
print(np.max(time_df[~max_mask]['time_ns']/time_df[~max_mask]['n_max']))
print(len(time_df[~max_mask]['time_ns']))
```
